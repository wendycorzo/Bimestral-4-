```python
import matplotlib.pyplot as plt
import numpy as np
import sys

def lineal():
  def funcion_lineal(m, x, b):
    return m*x+b
  N=100
  m= int(input("Digite el valor de la pendiente: "))
  b= int(input("Digite el valor del punto de corte con el eje y: "))

  x=np.linspace(-10,10, num= N)
  y= funcion_lineal(m,x,b)

  plt.plot(x, y, color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("Función Lineal: y=mx+b")
  plt.grid()
  plt.axhline(y=0, color="b")
  plt.axvline(x=0, color="b")
  plt.show()

  op=int(input("Desea continuar(1) o volver(2)? Digite:"))
  if op==1:
    lineal(m,x,b)
  elif op==2:
    menu()


def cuadr():
  def funcion_cuadr(a,b,c,x):
    return a*x**2+b*x+c
  
  N=100
  a= int(input("Digite el valor de a: "))
  b= int(input("Digite el valor del b: "))
  c= int(input("Digite el valor de c: "))

  x=np.linspace(-10,10, num= N)
  y= funcion_cuadr(a,b,c,x)

  plt.plot(x, y, color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("Función Cuadrática: y=ax²+bx+c")
  plt.grid()
  plt.axhline(y=0, color="b")
  plt.axvline(x=0, color="b")
  plt.show()

  op=int(input("Desea continuar(1) o volver(2)? Digite:"))
  if op==1:
    cuadr()
  elif op==2:
    menu()

def cub():
  def funcion_cub(a,b,c,d, x):
    return a*x**3+b*x**2+c*x+d

  N=100
  a= int(input("Digite el valor de a: "))
  b= int(input("Digite el valor del b: "))
  c= int(input("Digite el valor de c: "))
  d= int(input("Digite el valor de d: "))

  x=np.linspace(-10,10, num= N)
  y= funcion_cub(a,b,c,d, x)


  plt.plot(x, y, color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("Función Cúbica: y=ax³+bx²+cx+d")
  plt.grid()
  plt.axhline(y=0, color="b")
  plt.axvline(x=0, color="b")
  plt.show()

  op=int(input("¿Quieres continuar? (1) o volver(2)? Digite:"))
  if op==1:
    cub()
  elif op==2:
    menu()


def logt():
  N=100
  x=np.linspace(-10,10, num= N)
  y= np.log(x)

  plt.plot(x, y, color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("Función Logarítmica: y=log(x)")
  plt.grid()
  plt.axhline(y=0, color="b")
  plt.axvline(x=0, color="b")
  plt.show()


  op=int(input("Desea continuar(1) o volver(2)? Digite:"))
  if op==1:
    logt()
  elif op==2:
    menu()

def exp():
  N=100
  a=float(input("Digite el valor de a: "))
  x=np.linspace(-10,10, num= N)
  y= a**x

  plt.plot(x, y, color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("Función Exponencial: y=a^x")
  plt.grid()
  plt.axhline(y=0, color="b")
  plt.axvline(x=0, color="b")
  plt.show()

  op=int(input("Desea continuar(1) o volver(2)? Digite:"))
  if op==1:
    exp()
  elif op==2:
    menu()
def trig():
  op=int(input("\nDigite 1 para el Seno, 2 Para el Coseno, 3 Para volver a el menu . Digite: "))
  if op==1:
    x= np.linspace(-np.pi, np.pi, 256)
    y=np.sin(x)

    plt.plot(x,y)
    plt.ylim(-2,2)
    plt.show()

    trig()

  elif op==2:
    x= np.linspace(-np.pi, np.pi, 256)
    y=np.cos(x)

    plt.plot(x,y)
    plt.ylim(-2,2)
    plt.show()
    trig()
  elif op==3:
    menu()
  else:
    print("Error. Digite una de las dos opciones")




def menu():
  print("Gráficas:\n ")
  print("1 para lineal\n ")
  print("2 para cuadrática\n ")
  print("3 para cúbica\n  ")
  print("4 para logarítmica\n ")
  print("5 para exponencial\n ")
  print("6 para trigonométrica \n ")
  print("7 para cerrar el programa\n ")

  opcion=int(input("Digite la opción que quieres: "))

  if opcion==1:
    lineal()

  elif opcion==2:
    cuadr()

  elif opcion==3:
    cub()


  elif opcion==4:
    logt()


  elif opcion==5:
    exp()

  elif opcion==6:
    trig()

  elif opcion==7:
    sys.exit()


menu()
```
